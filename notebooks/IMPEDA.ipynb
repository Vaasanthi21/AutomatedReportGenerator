```python
# EDA.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/sample_data.csv", parse_dates=['Date'])

# Basic Info
print(df.info())
print(df.describe())

# Add a Profit column
df['Profit'] = df['Revenue'] - df['Expenses']

# Line plot of Revenue, Expenses, and Profit
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Revenue'], label='Revenue', marker='o')
plt.plot(df['Date'], df['Expenses'], label='Expenses', marker='o')
plt.plot(df['Date'], df['Profit'], label='Profit', marker='o')
plt.title('Revenue, Expenses, and Profit Over Time')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
